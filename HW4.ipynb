### Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = np.round(df.corr(), 2)
corr_matrix
```

```python
upp_mat = np.triu(df.corr())
```

```python
plt.figure(figsize = (15,11))
sns.heatmap(df.corr(), annot=True, cmap = 'viridis', mask = upp_mat);
```

```python
plt.figure(figsize=(15,8))
sns.lmplot(data=df, x="sqft_living", y="price", line_kws={'color': 'blue'})
plt.xlabel('Квадратура жилой площади, кв. фт')
plt.ylabel('Стоимость дома')
plt.title('Зависимость стоимости дома от квадратуры жилой площади', loc='left');
```

Выводы
Основное колличество проданных домов иммет площадь до 6 тыс кв фт и стоимость до 500 тыс долл.


```python
plt.figure(figsize=(9, 7))
sns.boxplot(data = df, x = 'grade', y = 'price')
plt.xlabel('Класс дома')
plt.ylabel('Стоимость дома')
plt.title('Зависимость стоимости дома от его класса');
```

#### Выводы
- стоимость домов 1-9 класса практически на одном уровне -до 100 тыс.долл,10-11 класса -до 200 тыс долл,12 класса- 180-300 тыс долл,13 -250 - 400 тыс.долл.Самый большой размах цен у дорогих домов 13 класса.



```python
plt.figure(figsize=size_)
sns.lmplot(data=df, x="sqft_above", y="price", line_kws={'color': 'blue'})
plt.xlabel('Квадратура общей площади дома, кв. фт')
plt.ylabel('Стоимость дома')
plt.title('Зависимость стоимости дома от общей квадратуры дома', loc='left'); 
```

Выводы

 большинство домов имеет площадь под застройку до 6 000 квадратных футов и стоимость до 400 000 долларов

```python
plt.figure(figsize=(12, 8))
sns.violinplot(data = df, x = 'waterfront', y = 'price')
plt.xlabel('Вид на набережную')
plt.ylabel('Стоимость дома')
plt.title('Распределение стоимости домов в зависимости от наличия вида на набережную');
```

Выводы

Большинство  домов не имеют вида на набережную, а стоимость таких домов относительно редко превышает стоимость в 250 000 долларов,стоимость домов на набережной  варируется в пределах до 600 000 долларов и распределение цен более равномерное;

```python
plt.figure(figsize=size_)
sns.histplot(data = df, x = 'yr_built', y = 'price', bins = 80)
plt.xlabel('Год постройки дома')
plt.ylabel('Стоимость дома')
plt.title('Распределение стоимости домов в зависимости от года его постройки');
```

Выводы

Основное колличество домов 1950-2020 годов,прослеживается явный ценовой разброс ,дома более новые,2000-2020 годов стоят дороже.
